```python
data = df
```

```python
plt.hist(data['Annual Return']*100, bins=10)

# add a title and axis labels
plt.title('Annual Returns')
plt.xlabel('Return (%)')
plt.ylabel('Frequency')

# display the chart
plt.show()
```

```python
bins = [0.0, 0.33, 0.6, 1.0]

# Define the bin labels
labels = ['Small cap', 'Medium cap', 'Large cap']
# Assign each value to a bin
data['Cap'] = pd.cut(data[' Large S/P '], bins=bins, labels=labels)
data['Cap'].fillna(value='Small cap', inplace=True)
cap_avg = data.groupby('Cap')[['Annual Return']].mean()
cap_avg['Annual Return'] =cap_avg['Annual Return']*100
cap_avg
```

```python
bins = [0.05,0.10,0.15,0.2]

# Define the bin labels
labels = ['Low','Medium','High']
# Assign each value to a bin
data['return'] = pd.cut(data['Annual Return'], bins=bins, labels=labels)
data.pivot_table('ID',columns=['return'],index = ['Cap'],aggfunc=['count'],margins=True)
```

```python
bins = [0.08,0.11,0.13,0.16]

# Define the bin labels
labels = ['Low','Medium','High']
# Assign each value to a bin
data['risk'] = pd.cut(data['Total Risk'], bins=bins, labels=labels)
data.pivot_table('ID',columns=['risk'],index = ['return'],aggfunc='count',margins=True)
```

```python
def get_sign(x):
    if x > 0:
        return 'Positive'
    elif x < 0:
        return 'Negative'
    else:
        return 'Zero'

# Apply the lambda function to create a new column 'sign'
data['sign'] = data['Excess Return'].apply(lambda x: get_sign(x))

# Display the resulting DataFrame
data.groupby('sign')[['ID']].count()
```
